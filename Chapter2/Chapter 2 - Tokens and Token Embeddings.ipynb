```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
# Load model and tokenizer
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct", #ÊåáÂÆöË¶ÅÂä†ËΩΩÁöÑÈ¢ÑËÆ≠ÁªÉÊ®°Âûã
    device_map="cuda", #ÊåáÂÆöÊ®°ÂûãËøêË°åÁöÑËÆæÂ§áÔºåÂ∞ÜÊ®°ÂûãÂä†ËΩΩÂà∞GPU‰∏äËøêË°åÔºàÈúÄË¶ÅNVIDIA GPUÂíåCUDAÔºâ
    torch_dtype="auto", #Ëá™Âä®ÈÄâÊã©Êï∞ÂÄºÁ≤æÂ∫¶
    trust_remote_code=False,
)
```

```python
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

```python
prompt = "Write an email apologizing to Sarah for the tragic gardening mishap. Explain how it happened.<|assistant|>"
#Ë¶ÅÊ±ÇÂÜô‰∏ÄÂ∞ÅÂêëSarahÈÅìÊ≠âÁöÑÈÇÆ‰ª∂ÔºåËß£ÈáäÂõ≠Ëâ∫‰∫ãÊïÖÁöÑÂèëÁîü

# Tokenize the input prompt
input_ids = tokenizer(prompt, #Â∞ÜÊñáÊú¨ËΩ¨Êç¢‰∏∫Ê®°ÂûãËÉΩÁêÜËß£ÁöÑÊï∞Â≠ótokenÂ∫èÂàó
            return_tensors="pt" #ËøîÂõûPyTorchÂº†ÈáèÔºàtensorÔºâ
            ).input_ids.to("cuda") #ÊèêÂèñÂàÜËØçÂêéÁöÑÊï∞Â≠óIDÂ∫èÂàó

# Generate the text
generation_output = model.generate( #Ë∞ÉÁî®Ê®°ÂûãÁöÑÁîüÊàêÊñπÊ≥ï
  input_ids=input_ids, #Êèê‰æõÂàÜËØçÂêéÁöÑËæìÂÖ•
  max_new_tokens=20 #ËæìÂá∫20‰∏™token
)

# Print the output
print(tokenizer.decode #Â∞ÜÊï∞Â≠ótokenÂ∫èÂàóËΩ¨Êç¢ÂõûÂèØËØªÊñáÊú¨
 (generation_output[0]))
```

# Comparing Trained LLM Tokenizers


```python
from transformers import AutoModelForCausalLM, AutoTokenizer

colors_list = [ #RGBÁºñÁ†ÅÔºåÂÖ≠‰∏™È¢úËâ≤Âæ™ÁéØÊâìÂç∞
    '102;194;165', '252;141;98', '141;160;203',
    '231;138;195', '166;216;84', '255;217;47'
]

def show_tokens(sentence, tokenizer_name):
    tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)
    token_ids = tokenizer(sentence).input_ids
    for idx, t in enumerate(token_ids):
        print(                # idx % len(colors_list)ÊòØÂèñ‰ΩôÊï∞Ôºå
                           # ‰∏çÊñ≠ËæìÂá∫0,1,2,3,4,5ÔºåÂæ™ÁéØÂèñËâ≤
            f'\x1b[0;30;48;2;{colors_list[idx % len(colors_list)]}m' +
            tokenizer.decode(t) +
            '\x1b[0m',
            end=' ' #ÊØè‰∏™token‰πãÂêéËæìÂá∫‰∏Ä‰∏™Á©∫Ê†º
        )
```

```python
text = """
English and CAPITALIZATION
üéµ È∏ü
show_tokens False None elif == >= else: two tabs:"    " Three tabs: "       "
12.0*50=600
"""
```

```python
show_tokens(text, "bert-base-uncased")
```

```python
show_tokens(text, "bert-base-cased")
```

```python
show_tokens(text, "gpt2")
```

```python
show_tokens(text, "google/flan-t5-small")
```

```python
# The official is `tiktoken` but this the same tokenizer on the HF platform
show_tokens(text, "Xenova/gpt-4")
```

```python
# You need to request access before being able to use this tokenizer
show_tokens(text, "bigcode/starcoder2-15b")
```

```python
show_tokens(text, "facebook/galactica-1.3b")
```

```python
show_tokens(text, "microsoft/Phi-3-mini-4k-instruct")
```

# Contextualized Word Embeddings From a Language Model (Like BERT)

```python
from transformers import AutoModel, AutoTokenizer

# ‰∏ãËΩΩÂàÜËØçÂô®
tokenizer = AutoTokenizer.from_pretrained("microsoft/deberta-base")

# ‰∏ãËΩΩËØ≠Ë®ÄÊ®°Âûã
model = AutoModel.from_pretrained("microsoft/deberta-v3-xsmall")

# ÂàÜËØçÂåñÔºå‰∏∫Âè•Â≠êÁîüÊàêtoken
tokens = tokenizer('Hello world', return_tensors='pt')

# Â∞ÜÂ§ÑÁêÜÂêéÁöÑ tokens ËæìÂÖ•Ê®°ÂûãÂπ∂Ëé∑ÂèñËæìÂá∫
output = model(**tokens)[0]
# Á¥¢Âºï[0]Áî®‰∫éËé∑ÂèñËøô‰∏™ÂÖÉÁªÑ‰∏≠ÁöÑÁ¨¨‰∏Ä‰∏™‰πüÊòØÊúÄÈáçË¶ÅÁöÑÂÖÉÁ¥†‚Äî‚ÄîÊúÄÂêé‰∏ÄÂ±ÇÈöêËóèÁä∂ÊÄÅ„ÄÇ
```

```python
output.shape
# Ëøô‰∏™ËæìÂá∫ÁöÑÂΩ¢Áä∂ÊòØ(batch_size, sequence_length, hidden_size)
# ÂØπ‰∫éÊú¨‰æãÔºåbatch_size ÊòØ1Ôºà‰∏Ä‰∏™Âè•Â≠êÔºâÔºåsequence_lengthÊòØËæìÂÖ•token
# ÁöÑÊï∞ÈáèÔºà4ÔºåÂåÖÊã¨ÁâπÊÆätokenÔºâÔºåhidden_sizeÊòØÊ®°ÂûãÁöÑÈöêËóèÂ±ÇÁª¥Â∫¶Ôºà384ÔºåÂèñÂÜ≥‰∫éÊ®°ÂûãÈÖçÁΩÆÔºâ„ÄÇ
```

# Recommending songs by embeddings

```python
import pandas as pd
from urllib import request

# Get the playlist dataset file
# Ëøô‰∏™txtÊñá‰ª∂ÂåÖÂê´‰∫ÜÂπøÊí≠ÁîµÂè∞ÁöÑÊí≠ÊîæÂàóË°®Êï∞ÊçÆ
data = request.urlopen('https://storage.googleapis.com/maps-premium/dataset/yes_complete/train.txt')

# Parse the playlist dataset file. Skip the first two lines as
# they only contain metadata
# Ë∑≥ËøáÂâç‰∏§Ë°åÔºåÊñá‰ª∂ÂºÄÂ§¥ÂèØËÉΩÂåÖÂê´‰ΩúËÄÖ„ÄÅÊó•ÊúüÁ≠âÂÖÉÊï∞ÊçÆ‰ø°ÊÅØÔºåËÄåÈùûÂÆûÈôÖÁöÑÊí≠ÊîæÂàóË°®Êï∞ÊçÆ„ÄÇÊ∏ÖÊ¥óÊï∞ÊçÆ
lines = data.read().decode("utf-8").split('\n')[2:]

# Ê†ºÂºèÂåñÊí≠ÊîæÂàóË°®ÔºåÂπ∂‰∏îÂè™‰øùÁïôÂåÖÂê´Â§ö‰∫é‰∏ÄÈ¶ñÊ≠åÊõ≤ÁöÑÊí≠ÊîæÂàóË°®
playlists = [s.rstrip().split() for s in lines if len(s.split()) > 1]

# ‰ªéÂè¶‰∏Ä‰∏™URL‰∏ãËΩΩsong_hash.txtÊñá‰ª∂
songs_file = request.urlopen('https://storage.googleapis.com/maps-premium/dataset/yes_complete/song_hash.txt')
songs_file = songs_file.read().decode("utf-8").split('\n')
songs = [s.rstrip().split('\t') for s in songs_file] # ÂàõÂª∫Ê≠åÊõ≤‰ø°ÊÅØ DataFrame
songs_df = pd.DataFrame(data=songs, columns = ['id', 'title', 'artist']) #Ê≠åÊõ≤ID„ÄÅÊ≠åÊõ≤Ê†áÈ¢ò„ÄÅËâ∫ÊúØÂÆ∂
songs_df = songs_df.set_index('id')
```

```python
print( 'Playlist #1:\n ', playlists[0], '\n')
print( 'Playlist #2:\n ', playlists[1])
```

```python
from gensim.models import Word2Vec

# Train our Word2Vec model
model = Word2Vec(
    playlists, vector_size=32, window=20, negative=50, min_count=1, workers=4
)
```

```python
song_id = 2172

# Ask the model for songs similar to song #2172
model.wv.most_similar(positive=str(song_id))
```

```python
print(songs_df.iloc[2172])
```

```python
import numpy as np

# Ê†πÊçÆËæìÂÖ•ÁöÑÊ≠åÊõ≤IDÔºå‰ΩøÁî®ËÆ≠ÁªÉÂ•ΩÁöÑword2vecÊ®°ÂûãÔºàmodelÔºâÊù•Êü•ÊâæÊúÄÁõ∏‰ºº
# ÁöÑ5È¶ñÊ≠åÊõ≤ÔºåÂπ∂ËøîÂõûËøô‰∫õÁõ∏‰ººÊ≠åÊõ≤ÁöÑËØ¶ÁªÜ‰ø°ÊÅØÔºàÊ≠åÂêçÂíåËâ∫ÊúØÂÆ∂Ôºâ
def print_recommendations(song_id):
    similar_songs = np.array(
        model.wv.most_similar(positive=str(song_id),topn=5)# Êü•ÊâæÊúÄÁõ∏‰ººÁöÑ‰∫îÈ¶ñÊ≠åÊõ≤
    )[:,0] # ÊèêÂèñÂá∫ÊâÄÊúâÁõ∏‰ººÊ≠åÊõ≤ÁöÑIDÔºå‰∏¢ÂºÉÁõ∏‰ººÂ∫¶ÂæóÂàÜÔºå
        # ÁªìÊûúÊòØ‰∏Ä‰∏™ÂåÖÂê´‰∫îÈ¶ñÊ≠åÊõ≤IDÁöÑ‰∏ÄÁª¥Êï∞ÁªÑ

    return  songs_df.iloc[similar_songs] # ËøîÂõûÁªìÊûúÊòØ‰∏Ä‰∏™DataFrameÔºå
                        # ÂåÖÂê´5È¶ñÊúÄÁõ∏‰ººÊ≠åÊõ≤ÁöÑID„ÄÅÊ†áÈ¢òÂíåËâ∫ÊúØÂÆ∂‰ø°ÊÅØ

# Extract recommendations
print_recommendations(2172)
```
